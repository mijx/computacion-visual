Importaciones:

```python
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen
original = cv2.imread('img/aurora.png')
```

```python
# Mostrar imagenes de cv2 con plt
def display(image):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```

## Imagen original

```python
display(original)
```

## Acceder y mostrar los canales RGB y HSV por separado

```python
# Convertir la imagen a RGB
rgb_image = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)

# Mostrar los canales RGB
r, g, b = cv2.split(rgb_image)

# Convertir la imagen a HSV
hsv_image = cv2.cvtColor(original, cv2.COLOR_BGR2HSV)

# Extraer los canales HSV
h, s, v = cv2.split(hsv_image)
```

### Canales RGB

```python
print("R:")
display(r)

print("G:")
display(g)  

print("B:")
display(b)
```

### Canales HSV

```python
print("H:")
display(h)

print("S:")
display(s)

print("V:")
display(v)
```

## Aplicaciones de Slicing de matrices

### Cambiar el color de un rea rectangular
Especficamente, se edita un rectngulo ubicado en la mitad de la imagen con dimensin 200x200 pxeles.

```python
slicingcolor = original.copy()
height, width, _ = original.shape

# Definir el rea rectangular
start_x = width // 4
start_y = height // 4
end_x = 3 * width // 4
end_y = 3 * height // 4

# Cambiar el color del rea seleccionada
slicingcolor[start_y:end_y, start_x:end_x] = [0, 0, 255]

display(slicingcolor)
```

### Sustituir una regin por otra parte de la imagen
Cambiamos la mitad derecha de la imagen por una copia de la mitad izquierda de la misma.

```python
slicingsustitucion = original.copy()
# Dividir la imagen en dos mitades
left_half = original[:, :width // 2]

# Intercambiar las mitades
slicingsustitucion = slicingsustitucion.copy()
slicingsustitucion[:, width // 2:] = left_half  # Sustituir la mitad derecha por la izquierda

# Mostrar la imagen modificada
display(slicingsustitucion)
```
